**Q2 : 4**

Let $T_1, T_2 \in SE(3) $ i.e. $T_i = \begin{bmatrix} R_i & p_i \\ 0 & 1 \end{bmatrix}$

Let $f(T_1, T_2, u) = T_1T_2\begin{bmatrix}u \\ 1 \end{bmatrix}$

where $u \in ℝ^3$

Derive the Jacobian of f with respect to orientation and translation parameters of T_1, T_2.


---
##Ans (do not delete this cell)
---


##10%
The Jacobian is a 3 x 12 matrix given as:

$J = \begin{bmatrix} J_{R_1} & J_{p_1} & J_{R_2} & J_{p_2}\end{bmatrix}$


##10%

Expanding the function f, we get:

$f(R_1, p_1, R_2, p_2, u) = R_1R_2u+R_1p_2+p_1$

##30%

The Jacobian w.r.t $R_1$ is

$J_{R_1} = -R_1[R_2u+p_2]_\times$

##10%

With respect to $p_1$ is

$J_{p_1} = I$

##30%

With respect to $R_2$ is

$J_{R_2} = -R_1R_2[u]_\times $

##10%
and finally

$J_{p_2} = R_1$

Hence putting all of these together, we get the 3x12 Jacobian.
